```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from matplotlib.colors import ListedColormap
```

```python
# Tạo dữ liệu mẫu
n_samples = 500
n_features = 2
n_clusters = 4
random_state = 42

X, y = make_blobs(n_samples=n_samples, n_features=n_features,
            centers=n_clusters, cluster_std=1.0,
            random_state=random_state)
```

```python
# Chạy thuật toán K-means
kmeans = KMeans(n_clusters=n_clusters, random_state=random_state, n_init=10)
kmeans.fit(X)
y_pred = kmeans.predict(X)
centers = kmeans.cluster_centers_
```

```python
# Vẽ kết quả
plt.figure(figsize=(10, 6))
```

```python
# Tạo bảng màu
cmap = ListedColormap(['#FF5733', '#33FF57', '#3357FF', '#F3FF33'])
```

```python
# Vẽ các điểm dữ liệu
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap=cmap, alpha=0.7, s=50)
```

```python
# Vẽ các tâm cụm
plt.scatter(centers[:, 0], centers[:, 1], c=range(n_clusters), cmap=cmap,
edgecolor='black', s=200, alpha=0.9, marker='X')
```

```python
# Vẽ các tâm cụm
plt.scatter(centers[:, 0], centers[:, 1], c=range(n_clusters), cmap=cmap,
        edgecolor='black', s=200, alpha=0.9, marker='X')

plt.title('K-means Clustering with scikit-learn')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Vẽ đồ thị Elbow để xác định K tối ưu
wcss = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=random_state, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_range, wcss, marker='o', linestyle='-')
plt.title('Phương pháp Elbow cho việc xác định K tối ưu')
plt.xlabel('Số lượng cụm (K)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
